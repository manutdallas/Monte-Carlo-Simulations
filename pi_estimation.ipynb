```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
   inside : list to store the points falling inside a circle centered at (0,0) with unit radius

  outside : list to store the points falling outside a circle centered at (0,0) with unit radius
            and inside square with center at origin and side of 2 sq. units.

      est : list to store estimated valued of pi after every iteration.

        N : number of iterations
"""
inside = []
outside = []
est = []
N = 60000
cnt = 0
for i in range(N):
  x = np.random.uniform(-1,1)
  y = np.random.uniform(-1,1)
  if x*x + y*y <= 1:
    inside.append([x,y])
    cnt = cnt + 1
  else:
    outside.append([x,y])
  est.append(4.0 * (cnt/(i+1)))
pi = 4.0 * (cnt/N)
print("Estimated Value of pi :: ", pi)
plt.scatter([x[0] for x in inside], [x[1] for x in inside], marker='.', alpha=0.6)
plt.scatter([x[0] for x in outside], [x[1] for x in outside], marker='*', alpha=0.2)
plt.show()
plt.plot(est, label='estimated value of pi')
plt.axhline(3.14159265358, color='r', linestyle='-', label='true value of pi')
plt.legend()
plt.show()
```
